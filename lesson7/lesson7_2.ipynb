```python
import os
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.embeddings import FakeEmbeddings
from langchain_chroma import Chroma
import chromadb

current_dir = os.path.dirname(os.path.abspath("__file__"))
persistent_directory = os.path.join(current_dir, "db", "chroma_db_v3")
client = chromadb.PersistentClient(path=persistent_directory)

# 使用 FakeEmbeddings 進行測試（不需要 API Key）
# 注意：使用 1024 維度以匹配已存在的資料庫
embeddings = FakeEmbeddings(size=1024)

db = Chroma(
    client = client,
    embedding_function=embeddings,
    collection_name="smartphone_manual"
)

retriever = db.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}
)

query = "如何設定指紋辨識?"

relevant_docs = retriever.invoke(query)
for doc in relevant_docs:
    print(doc.page_content)
    print("=" * 60)
```

```python
# RAG 鏈 - 使用 Ollama 和向量搜尋
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 創建提示模板
prompt = ChatPromptTemplate.from_template("""
基於以下上下文回答問題。請仔細閱讀上下文內容，並提供準確且詳細的回答。

上下文:
{context}

問題: {question}

請提供詳細且有用的回答:
""")

# 使用正確的 Ollama 模型名稱
llm = ChatOllama(model="gemma3:1b")

# 創建 RAG 鏈
rag_chain = (
    prompt
    | llm
    | StrOutputParser()
)

# 測試 RAG 鏈 - 使用有相關資訊的查詢
query = "如何設定臉部辨識?"

# 獲取相關文件
relevant_docs = retriever.invoke(query)
context = "\n\n".join([doc.page_content for doc in relevant_docs])

print(f"找到 {len(relevant_docs)} 個相關文件")
print("上下文長度:", len(context))

# 使用 RAG 鏈回答問題
result = rag_chain.invoke({
    "question": query,
    "context": context
})

print("\n問題:", query)
print("\n回答:")
print(result)
```
